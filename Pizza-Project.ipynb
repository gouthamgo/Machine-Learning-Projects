```python
import pandas as pd

# OUR "REAL" PIZZA DATA
pizza_data = {
    'pepperoni': [1, 1, 0, 0, 1],  # 1 = likes, 0 = hates
    'mushrooms': [0, 1, 1, 1, 0],
    'sausage':   [1, 0, 0, 0, 1],
    'anchovies': [0, 0, 1, 1, 0],
    'predict':   [1, 0, 1, 1, 1]
}

pizza = pd.DataFrame(pizza_data)

print("My  data:")
print(pizza)
```

```python
from sklearn.tree import DecisionTreeClassifier

X = pizza[['pepperoni', 'mushrooms', 'anchovies']]  # Clues (what they like)
y = pizza['predict']                               # Answer (will they like sausage?)

pattern_spotter = DecisionTreeClassifier()
pattern_spotter.fit(X, y)  # "Learn patterns from my pizza notes"
```

```python
import numpy as np
import pandas as pd

# NEW FRIEND: Likes pepperoni, hates mushrooms, hates anchovies
new_friend_data = {'pepperoni': [1], 'mushrooms': [0], 'anchovies': [0]}  # [pepperoni, mushrooms, anchovies]
new_friend_df = pd.DataFrame(new_friend_data)

# ASK THE PATTERN-SPOTTER
prediction = pattern_spotter.predict(new_friend_df)

print(f"\nNew friend: pepperoni=1, mushrooms=0, anchovies=0")
print(f"Prediction: {'LIKES sausage' if prediction[0] == 1 else 'HATES sausage'}")
```

```python
from sklearn import tree
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
tree.plot_tree(
    pattern_spotter,
    feature_names=['pepperoni', 'mushrooms', 'anchovies'],
    class_names=['hates sausage', 'likes sausage'],
    filled=True
)
plt.title("How the pattern-spotter thinks")
plt.show()
```
